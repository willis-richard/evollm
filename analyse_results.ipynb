```python
import pandas as pd

df = pd.read_csv("./results_full.txt")

df = df.drop(columns=['Interaction index', 'Player index', 'Opponent index', 'Repetition', 'Turns', 'Score per turn', 'Score difference per turn', 'Win', 'CC count', 'CD count',
       'DC count', 'DD count'])
```

```python
df.head()
```

```python
mean_scores = df.groupby('Player name')['Score'].mean()
ranked_df = pd.DataFrame(mean_scores).rename(columns={'Score': 'Mean Score'})
ranked_df
```

```python
df.pivot_table(index='Player name', columns='Opponent name', values='Score', aggfunc='mean')
```
